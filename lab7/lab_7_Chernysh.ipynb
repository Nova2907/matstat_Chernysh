```python
# –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞
def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])

def getRootVal(root):
    return root[0]

def setRootVal(root, newVal):
    root[0] = newVal

def getLeftChild(root):
    return root[1]

def getRightChild(root):
    return root[2]

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≤–∏–≤–æ–¥—É –¥–µ—Ä–µ–≤–∞ —É –≤–∏–≥–ª—è–¥—ñ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –¥–µ—Ä–µ–≤–∞
def printTree(tree, level=0):
    if tree:
        printTree(getRightChild(tree), level + 1)
        print(" " * 4 * level + "->", getRootVal(tree))
        printTree(getLeftChild(tree), level + 1)

# –ü—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –¥–µ—Ä–µ–≤–∞
tree = BinaryTree(7)
insertLeft(tree, 6)
insertLeft(getLeftChild(tree), 3)
insertRight(tree, 9)
insertRight(getRightChild(tree), 13)
insertRight(getRightChild(getRightChild(tree)), 21)
insertLeft(getRightChild(getRightChild(getRightChild(tree))), 33)
insertRight(getRightChild(getRightChild(getRightChild(tree))), 52)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –¥–µ—Ä–µ–≤–∞
print("–°—Ç–≤–æ—Ä–µ–Ω–µ –¥–µ—Ä–µ–≤–æ:")
printTree(tree)

```

```python
# –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞
def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])

def getRootVal(root):
    return root[0]

def setRootVal(root, newVal):
    root[0] = newVal

def getLeftChild(root):
    return root[1]

def getRightChild(root):
    return root[2]

def setLeftChild(root, newChild):
    root[1] = newChild

def setRightChild(root, newChild):
    root[2] = newChild

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –∑–∞–¥–∞–Ω–æ—ó –≥—ñ–ª–∫–∏
def deleteBranch(root, value):
    """–í–∏–¥–∞–ª—è—î –≥—ñ–ª–∫—É –¥–µ—Ä–µ–≤–∞ –∑ –∑–∞–¥–∞–Ω–∏–º –∑–Ω–∞—á–µ–Ω–Ω—è–º"""
    if not root:
        return False

    # –Ø–∫—â–æ –∑–Ω–∞—á–µ–Ω–Ω—è –∑–Ω–∞–π–¥–µ–Ω–æ –≤ –ª—ñ–≤—ñ–π –≥—ñ–ª—Ü—ñ, –≤–∏–¥–∞–ª—è—î–º–æ –ª—ñ–≤—É –ø—ñ–¥–≥—ñ–ª–∫—É
    if getLeftChild(root) and getRootVal(getLeftChild(root)) == value:
        setLeftChild(root, [])
        return True

    # –Ø–∫—â–æ –∑–Ω–∞—á–µ–Ω–Ω—è –∑–Ω–∞–π–¥–µ–Ω–æ –≤ –ø—Ä–∞–≤—ñ–π –≥—ñ–ª—Ü—ñ, –≤–∏–¥–∞–ª—è—î–º–æ –ø—Ä–∞–≤—É –ø—ñ–¥–≥—ñ–ª–∫—É
    if getRightChild(root) and getRootVal(getRightChild(root)) == value:
        setRightChild(root, [])
        return True

    # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ —à—É–∫–∞—î–º–æ –≤ –ª—ñ–≤—ñ–π —ñ –ø—Ä–∞–≤—ñ–π –ø—ñ–¥–≥—ñ–ª–∫–∞—Ö
    return deleteBranch(getLeftChild(root), value) or deleteBranch(getRightChild(root), value)

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≤–∏–≤–æ–¥—É –¥–µ—Ä–µ–≤–∞ —É –≤–∏–≥–ª—è–¥—ñ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –¥–µ—Ä–µ–≤–∞
def printTree(tree, level=0):
    if tree:
        printTree(getRightChild(tree), level + 1)
        print(" " * 4 * level + "->", getRootVal(tree))
        printTree(getLeftChild(tree), level + 1)

# –ü—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –¥–µ—Ä–µ–≤–∞
tree = BinaryTree(7)
insertLeft(tree, 6)
insertLeft(getLeftChild(tree), 3)
insertRight(tree, 9)
insertRight(getRightChild(tree), 13)
insertRight(getRightChild(getRightChild(tree)), 21)
insertLeft(getRightChild(getRightChild(getRightChild(tree))), 33)
insertRight(getRightChild(getRightChild(getRightChild(tree))), 52)

print("–î–µ—Ä–µ–≤–æ –¥–æ –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≥—ñ–ª–∫–∏:")
printTree(tree)

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –≥—ñ–ª–∫–∏ –∑ –∫–æ—Ä–µ–Ω–µ–º 21
deleteBranch(tree, 21)

print("\n–î–µ—Ä–µ–≤–æ –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≥—ñ–ª–∫–∏ –∑ –∫–æ—Ä–µ–Ω–µ–º 21:")
printTree(tree)

```

#### 2. **–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ—ó —Ä–æ–±–æ—Ç–∏**
   
1. –ù–∞–ø–∏—Å–∞—Ç–∏ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –∫—É–ø–∏ –∑ –±—É–¥—å-—è–∫–æ–≥–æ —Ä–∞–Ω–¥–æ–º–Ω–æ–≥–æ –º–∞—Å–∏–≤–∞.  
2. –î–æ–¥–∞—Ç–∏ –¥–æ –Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç, —è–∫–∏–π –¥–æ—Ä—ñ–≤–Ω—é—î –≤–∞—à–æ–º—É –ø–æ—Ä—è–¥–∫–æ–≤–æ–º—É –Ω–æ–º–µ—Ä—É —É —Å–ø–∏—Å–∫—É –≥—Ä—É–ø–∏.
3. –í–∏–ª—É—á–∏—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑ –∫—É–ø–∏.
4. –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å (—É —Å–µ—Ä–µ–¥–Ω—å–æ–º—É —ñ –≤ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É) –ø—Ä–æ—Ü–µ–¥—É—Ä search, insert —ñ delete —Ä–æ–±–æ—Ç–∏ –∑ –∫—É–ø–æ—é.

```python
import heapq
import random


def build_max_heap(arr):
    arr = [-x for x in arr]
    heapq.heapify(arr)
    return arr

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞ –¥–æ –∫—É–ø–∏ (20)
def insert_to_heap(heap, value):
    # –î–æ–¥–∞—î–º–æ –µ–ª–µ–º–µ–Ω—Ç, –∑–º—ñ–Ω—é—é—á–∏ –π–æ–≥–æ –∑–Ω–∞–∫ –Ω–∞ –≤—ñ–¥'—î–º–Ω–∏–π
    heapq.heappush(heap, -value)

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –∑ –∫—É–ø–∏
def delete_max_from_heap(heap):
    # –í–∏–¥–∞–ª—è—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç, –∑–º—ñ–Ω—é—é—á–∏ –π–æ–≥–æ –∑–Ω–∞–∫ –Ω–∞ –≤—ñ–¥'—î–º–Ω–∏–π
    max_value = -heapq.heappop(heap)
    return max_value

random_array = random.sample(range(1, 100), 10)  # –†–∞–Ω–¥–æ–º–Ω–∏–π –º–∞—Å–∏–≤
print("–†–∞–Ω–¥–æ–º–Ω–∏–π –º–∞—Å–∏–≤:", random_array)

max_heap = build_max_heap(random_array)  # –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –∫—É–ø–∏
print("max –∫—É–ø–∞:", [-x for x in max_heap])  # –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è –Ω–∞–∑–∞–¥ –≤ –ø–æ–∑–∏—Ç–∏–≤–Ω—ñ

# –î–æ–¥–∞—î–º–æ –µ–ª–µ–º–µ–Ω—Ç, —è–∫–∏–π –¥–æ—Ä—ñ–≤–Ω—é—î 20 
insert_to_heap(max_heap, 20)
print("–î–æ–¥–∞—î–º–æ –µ–ª–µ–º–µ–Ω—Ç 20:", [-x for x in max_heap])

# –í–∏–¥–∞–ª—è—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç
max_element = delete_max_from_heap(max_heap)
print("–í–∏–¥–∞–ª—è—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç:", max_element)
print("–ö—É–ø–∞ –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è:", [-x for x in max_heap])
```

    ‚Äì –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –ø—Ä–∞—Ü–µ–∑–¥–∞—Ç–Ω—ñ—Å—Ç—å –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä—ñ–∑–Ω–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.

   ‚Äì –≤–∏–º—ñ—Ä—è—Ç–∏ —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –æ—Å–Ω–æ–≤–Ω–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π (–ø–æ—à—É–∫, –≤—Å—Ç–∞–≤–∫–∞, –≤–∏–¥–∞–ª–µ–Ω–Ω—è) –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –¥–∞–Ω–∏—Ö.

   ‚Äì –ø–æ—Ä—ñ–≤–Ω—è—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –¥–∞–Ω–∏—Ö.

2. **–†–µ–∞–ª—ñ–∑—É–π—Ç–µ –≥–µ—à-—Ç–∞–±–ª–∏—Ü—é –∑ –ª–∞–Ω—Ü—é–∂–∫–æ–≤–∏–º –≥–µ—à—É–≤–∞–Ω–Ω—è–º.**

```python
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ç–∞–±–ª–∏—Ü—ñ –∑ –ø–æ—Ä–æ–∂–Ω—ñ–º–∏ —Å–ø–∏—Å–∫–∞–º–∏

    def hash_function(self, key):
        """–•–µ—à-—Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —ñ–Ω–¥–µ–∫—Å—É –≤ —Ç–∞–±–ª–∏—Ü—ñ."""
        return hash(key) % self.size

    def insert(self, key, value):
        """–í—Å—Ç–∞–≤–∫–∞ –µ–ª–µ–º–µ–Ω—Ç–∞ –≤ —Ç–∞–±–ª–∏—Ü—é."""
        index = self.hash_function(key)
        # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞, —á–∏ –≤–∂–µ —î —Ç–∞–∫–∏–π –∫–ª—é—á –≤ —Ç–∞–±–ª–∏—Ü—ñ, —è–∫—â–æ —î - –æ–Ω–æ–≤–ª—é—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è
        for idx, element in enumerate(self.table[index]):
            if element[0] == key:
                self.table[index][idx] = (key, value)  # –û–Ω–æ–≤–ª–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è
                return
        # –î–æ–¥–∞—î–º–æ –Ω–æ–≤—É –ø–∞—Ä—É (–∫–ª—é—á, –∑–Ω–∞—á–µ–Ω–Ω—è)
        self.table[index].append((key, value))

    def search(self, key):
        """–ü–æ—à—É–∫ –µ–ª–µ–º–µ–Ω—Ç–∞ –≤ —Ç–∞–±–ª–∏—Ü—ñ."""
        index = self.hash_function(key)
        # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –≤ —Å–ø–∏—Å–∫—É –∑–∞ —ñ–Ω–¥–µ–∫—Å–æ–º
        for element in self.table[index]:
            if element[0] == key:
                return element[1]
        return None

    def delete(self, key):
        """–í–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞ –∑ —Ç–∞–±–ª–∏—Ü—ñ."""
        index = self.hash_function(key)
        for i, element in enumerate(self.table[index]):
            if element[0] == key:
                del self.table[index][i]
                return True
        return False

    def __str__(self):
        
        result = "–≥–µ—à-—Ç–∞–±–ª–∏—Ü—è –∑ –ª–∞–Ω—Ü—é–∂–∫–æ–≤–∏–º –≥–µ—à—É–≤–∞–Ω–Ω—è–º.:\n"
        for i, bucket in enumerate(self.table):
            result += f"Index {i}: "
            if bucket:
                result += " -> ".join([f"({key}, {value})" for key, value in bucket]) + "\n"
            else:
                result += "Empty\n"
        return result


# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è:
ht = HashTable()

# –í—Å—Ç–∞–≤–∫–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
ht.insert(7, "value1")
ht.insert("test", "value2")
ht.insert(13, "value3")
ht.insert(21, "value4")

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ö–µ—à-—Ç–∞–±–ª–∏—Ü—ñ
print(ht)

# –ü–æ—à—É–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
print(f"–ü–æ—à—É–∫ 7: {ht.search(7)}")  # value1
print(f"–ü–æ—à—É–∫ 'test': {ht.search('test')}")  # value2

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
ht.delete(7)
print("–ü—ñ—Å–ª—è –í–∏–¥–∞–ª–µ–Ω–Ω—è  7:")
print(ht)
```

3. **–ü—Ä–æ–≤–µ–¥—ñ—Ç—å —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä—ñ–∑–Ω–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö (—Ü—ñ–ª—ñ —á–∏—Å–ª–∞, —Ä—è–¥–∫–∏, —Å–ø–∏—Å–∫–∏, —Å–ª–æ–≤–Ω–∏–∫–∏, –æ–±‚Äô—î–∫—Ç–∏):**
    1. –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –ø—Ä–∞—Ü–µ–∑–¥–∞—Ç–Ω—ñ—Å—Ç—å –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä—ñ–∑–Ω–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.
    2. –≤–∏–º—ñ—Ä—è—Ç–∏ —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –æ—Å–Ω–æ–≤–Ω–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π (–ø–æ—à—É–∫, –≤—Å—Ç–∞–≤–∫–∞, –≤–∏–¥–∞–ª–µ–Ω–Ω—è) –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –¥–∞–Ω–∏—Ö.

```python
import time

# –ö–ª–∞—Å –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for idx, element in enumerate(self.table[index]):
            if element[0] == key:
                self.table[index][idx] = (key, value)
                return
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for element in self.table[index]:
            if element[0] == key:
                return element[1]
        return None

    def delete(self, key):
        index = self.hash_function(key)
        for i, element in enumerate(self.table[index]):
            if element[0] == key:
                del self.table[index][i]
                return True
        return False

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –¥–∞–Ω–∏—Ö
def test_hash_table_operations(data_type):
    ht = HashTable()
    data = data_type()
    
    # –í—Å—Ç–∞–≤–∫–∞
    start_time = time.time()
    for i in range(1000):
        ht.insert(i, f"value_{i}")
    insert_time = time.time() - start_time
    
    # –ü–æ—à—É–∫
    start_time = time.time()
    for i in range(1000):
        ht.search(i)
    search_time = time.time() - start_time
    
    # –í–∏–¥–∞–ª–µ–Ω–Ω—è
    start_time = time.time()
    for i in range(1000):
        ht.delete(i)
    delete_time = time.time() - start_time
    
    return insert_time, search_time, delete_time

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –¥–∞–Ω–∏—Ö
data_types = [
    (int, "–¶—ñ–ª—ñ —á–∏—Å–ª–∞"),
    (str, "–†—è–¥–∫–∏"),
    (list, "–°–ø–∏—Å–∫–∏"),
    (dict, "–°–ª–æ–≤–Ω–∏–∫–∏"),
    (object, "–û–±'—î–∫—Ç–∏")
]

# –ü—Ä–æ–≤–µ–¥–µ–º–æ —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è
for data_type, name in data_types:
    insert_time, search_time, delete_time = test_hash_table_operations(data_type)
    print(f"–¢–µ—Å—Ç –¥–ª—è —Ç–∏–ø—É: {name}")
    print(f"–í—Å—Ç–∞–≤–∫–∞: {insert_time:.6f} —Å–µ–∫—É–Ω–¥")
    print(f"–ü–æ—à—É–∫: {search_time:.6f} —Å–µ–∫—É–Ω–¥")
    print(f"–í–∏–¥–∞–ª–µ–Ω–Ω—è: {delete_time:.6f} —Å–µ–∫—É–Ω–¥")
    print("-" * 40)
```
