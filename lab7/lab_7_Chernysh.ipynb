#### Бінарне дерево

```python
# Реалізація бінарного дерева
def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])

def getRootVal(root):
    return root[0]

def setRootVal(root, newVal):
    root[0] = newVal

def getLeftChild(root):
    return root[1]

def getRightChild(root):
    return root[2]

# Функція для виводу дерева у вигляді текстового дерева
def printTree(tree, level=0):
    if tree:
        printTree(getRightChild(tree), level + 1)
        print(" " * 4 * level + "->", getRootVal(tree))
        printTree(getLeftChild(tree), level + 1)

# Приклад створення дерева
tree = BinaryTree(7)
insertLeft(tree, 6)
insertLeft(getLeftChild(tree), 3)
insertRight(tree, 9)
insertRight(getRightChild(tree), 13)
insertRight(getRightChild(getRightChild(tree)), 21)
insertLeft(getRightChild(getRightChild(getRightChild(tree))), 33)
insertRight(getRightChild(getRightChild(getRightChild(tree))), 52)

# Виведення дерева
print("Створене дерево:")
printTree(tree)

```

#### Процедуру видалення заданої гілки дерева.

```python
# Функція для видалення заданої гілки
def deleteBranch(root, value):
    """Видаляє гілку дерева з заданим значенням"""
    if not root:
        return False

    # Якщо значення знайдено в лівій гілці, видаляємо ліву підгілку
    if getLeftChild(root) and getRootVal(getLeftChild(root)) == value:
        setLeftChild(root, [])
        return True

    # Якщо значення знайдено в правій гілці, видаляємо праву підгілку
    if getRightChild(root) and getRootVal(getRightChild(root)) == value:
        setRightChild(root, [])
        return True

    # Рекурсивно шукаємо в лівій і правій підгілках
    return deleteBranch(getLeftChild(root), value) or deleteBranch(getRightChild(root), value)

# Використання функції видалення
deleteBranch(tree, 21)

# Виведення дерева після видалення
print("Дерево після видалення гілки з коренем 21:")
printTree(tree)

```

```python
import heapq
import random


def build_max_heap(arr):
    arr = [-x for x in arr]
    heapq.heapify(arr)
    return arr

# Додавання елемента до купи (20)
def insert_to_heap(heap, value):
    # Додаємо елемент, змінюючи його знак на від'ємний
    heapq.heappush(heap, -value)

# Видалення максимального елемента з купи
def delete_max_from_heap(heap):
    # Видаляємо максимальний елемент, змінюючи його знак на від'ємний
    max_value = -heapq.heappop(heap)
    return max_value

random_array = random.sample(range(1, 100), 10)  # Рандомний масив
print("Рандомний масив:", random_array)

max_heap = build_max_heap(random_array)  # Створення купи
print("max купа:", [-x for x in max_heap])  # Перетворюємо значення назад в позитивні

# Додаємо елемент, який дорівнює 20 
insert_to_heap(max_heap, 20)
print("Додаємо елемент 20:", [-x for x in max_heap])

# Видаляємо максимальний елемент
max_element = delete_max_from_heap(max_heap)
print("Видаляємо максимальний елемент:", max_element)
print("Купа після видалення:", [-x for x in max_heap])
```

#### Оцінити асимптотичну складність (у середньому і в найгіршому випадку) процедур search, insert і delete роботи з купою.

1. Search 
У купі немає прямого способу ефективного пошуку довільного елемента, оскільки вона організована для швидкого доступу лише до кореня. Для загального випадку доведеться обійти всі елементи.

Середній випадок: 
O(n), де 𝑛 — кількість елементів.

Найгірший випадок: 
O(n), якщо шуканий елемент знаходиться в останньому вузлі.

2. Insert )
Під час вставки нового елемента: Елемент додається в кінець купи.
Проводиться "просіювання вгору" (heapify up), щоб відновити властивість купи.
У гіршому випадку елемент "підніметься" до кореня.

Середній випадок: O(logn), оскільки глибина купи ℎ=⌊log2𝑛⌋h=⌊log2​n⌋.

Найгірший випадок: O(logn).

3. Delete 
При видаленні:Кореневий елемент замінюється останнім елементом купи.
Останній елемент видаляється.
Виконується "просіювання вниз" (heapify down), щоб відновити властивість купи.
У гіршому випадку елемент опуститься до листа.

Середній випадок: 
O(logn).

Найгірший випадок: 
O(logn).

2. **Реалізуйте геш-таблицю з ланцюжковим гешуванням.**

```python
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # Ініціалізація таблиці з порожніми списками

    def hash_function(self, key):
        """Хеш-функція для обчислення індексу в таблиці."""
        return hash(key) % self.size

    def insert(self, key, value):
        """Вставка елемента в таблицю."""
        index = self.hash_function(key)
        # Перевірка, чи вже є такий ключ в таблиці, якщо є - оновлюємо значення
        for idx, element in enumerate(self.table[index]):
            if element[0] == key:
                self.table[index][idx] = (key, value)  # Оновлення значення
                return
        # Додаємо нову пару (ключ, значення)
        self.table[index].append((key, value))

    def search(self, key):
        """Пошук елемента в таблиці."""
        index = self.hash_function(key)
        # Перевірка кожного елемента в списку за індексом
        for element in self.table[index]:
            if element[0] == key:
                return element[1]
        return None

    def delete(self, key):
        """Видалення елемента з таблиці."""
        index = self.hash_function(key)
        for i, element in enumerate(self.table[index]):
            if element[0] == key:
                del self.table[index][i]
                return True
        return False

    def __str__(self):
        
        result = "геш-таблиця з ланцюжковим гешуванням.:\n"
        for i, bucket in enumerate(self.table):
            result += f"Index {i}: "
            if bucket:
                result += " -> ".join([f"({key}, {value})" for key, value in bucket]) + "\n"
            else:
                result += "Empty\n"
        return result


# Приклад використання:
ht = HashTable()

# Вставка елементів
ht.insert(7, "value1")
ht.insert("test", "value2")
ht.insert(13, "value3")
ht.insert(21, "value4")

# Виведення хеш-таблиці
print(ht)

# Пошук елементів
print(f"Пошук 7: {ht.search(7)}")  # value1
print(f"Пошук 'test': {ht.search('test')}")  # value2

# Видалення елементів
ht.delete(7)
print("Після Видалення  7:")
print(ht)
```

```python
import time

# Клас геш-таблиці
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for idx, element in enumerate(self.table[index]):
            if element[0] == key:
                self.table[index][idx] = (key, value)
                return
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for element in self.table[index]:
            if element[0] == key:
                return element[1]
        return None

    def delete(self, key):
        index = self.hash_function(key)
        for i, element in enumerate(self.table[index]):
            if element[0] == key:
                del self.table[index][i]
                return True
        return False

# Тестування для різних типів даних
def test_hash_table_operations(data_type):
    ht = HashTable()
    data = data_type()
    
    # Вставка
    start_time = time.time()
    for i in range(1000):
        ht.insert(i, f"value_{i}")
    insert_time = time.time() - start_time
    
    # Пошук
    start_time = time.time()
    for i in range(1000):
        ht.search(i)
    search_time = time.time() - start_time
    
    # Видалення
    start_time = time.time()
    for i in range(1000):
        ht.delete(i)
    delete_time = time.time() - start_time
    
    return insert_time, search_time, delete_time

# Тестування для різних типів даних
data_types = [
    (int, "Цілі числа"),
    (str, "Рядки"),
    (list, "Списки"),
    (dict, "Словники"),
    (object, "Об'єкти")
]

# Проведемо тестування
for data_type, name in data_types:
    insert_time, search_time, delete_time = test_hash_table_operations(data_type)
    print(f"Тест для типу: {name}")
    print(f"Вставка: {insert_time:.6f} секунд")
    print(f"Пошук: {search_time:.6f} секунд")
    print(f"Видалення: {delete_time:.6f} секунд")
    print("-" * 40)
```

### **Контрольні питання**

1. **Чим відрізняється структура бінарне дерево він бінарного дерева пошуку?**

Бінарне дерево: Кожен вузол має два нащадки (ліву та праву гілки), але немає певних правил щодо порядку елементів.
Бінарне дерево пошуку (BST): Має додаткові обмеження: всі елементи в лівому піддереві вузла менші за його значення, а в правому піддереві — більші.

2. **Чим відрізняється структура бінарне дерево від бінарної купи?**

Бінарне дерево: Просто структура з двома піддеревами для кожного вузла, без жорстких правил для значень.
Бінарна купа: Це спеціалізоване дерево, яке зберігає елементи таким чином, щоб виконувалася властивість купи (максимальна чи мінімальна), тобто в кожному вузлі значення більше (для максимального дерева) або менше (для мінімального дерева), ніж у його нащадках.

3. **Які існують типи дерев? Опишіть їхні основні характеристики та переваги.**

Бінарне дерево: Кожен вузол має два нащадки.
Бінарне дерево пошуку (BST): Відповідає певному порядку елементів.
Червонокоричневе дерево: Самобалансоване дерево з кольоровими вузлами для забезпечення рівноваги.
AVL-дерево: Самобалансоване дерево, яке підтримує баланс висоти піддерев.
Купа: Структура даних, що підтримує властивості мінімальної чи максимальної купи.
