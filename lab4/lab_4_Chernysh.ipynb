**Контрольні приклади:**

```python
def factorial_iterative(n):
    factorial = 1
    for i in range(1, n+1):
        factorial *= i
    return factorial

# Приклад використання
n = 5
print(f"Факторіал числа {n} через цикл: {factorial_iterative(n)}")
```

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Приклад використання
n = 5
print(f"Факторіал числа {n} через рекурсію: {factorial_recursive(n)}")
```

```python
def fibonacci(n):
    # Рекурсивне обчислення n-го числа Фібоначчі
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Приклад використання
print(fibonacci(30))

```

```python
def merge(left, right):
    # Зливає два відсортованих масиви left і right у один
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(lst):
    # Рекурсивна функція сортування з використанням попередньої функції
    if len(lst) < 2:
        return lst
    middle = len(lst) // 2
    left = mergesort(lst[:middle])
    right = mergesort(lst[middle:])
    return merge(left, right)

# Приклад масиву для сортування
a = [6, 5, 3, 1, 8, 7, 2, 4]

# Тестування злиття двох масивів
print(merge([2, 4, 5, 7], [1, 2, 3, 6]))  # Виведе: [1, 2, 2, 3, 4, 5, 6, 7]

# Тестування сортування масиву
print(mergesort(a))  # Виведе: [1, 2, 3, 4, 5, 6, 7, 8]
```

**Завдання для самостійної роботи. Реалізувати процедуру обчислення
факторіалу за допомогою циклу for. Оцінити асимптотичну складність
алгоритму.**

```python
def fac_for(n):
    # Обчислення факторіала за допомогою циклу for
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

# Приклад обчислення факторіала
n = 7
print(f"Факторіал {n}!: {fac_for(n)}")
```

**Завдання для самостійної роботи. Оцінити асимптотичну складність
рекурсивного алгоритму обчислення факторіалу.**

```python
def fac_recursive(n):
    # Рекурсивне обчислення факторіала
    if n == 0:
        return 1
    return n * fac_recursive(n - 1)

# Приклад обчислення факторіала
n = 7
print(f"Факторіал {n}! (рекурсивно): {fac_recursive(n)}")
```

**Завдання для самостійної роботи. Оцінити асимптотичну складність
рекурсивного алгоритму обчислення n-го числа Фібоначчі.**

```python
def fibonacci(n):
    # Рекурсивне обчислення n-го числа Фібоначчі
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Приклад обчислення числа Фібоначчі
n = 7
print(f"Число Фібоначчі {n}: {fibonacci(n)}")
```

**Завдання для самостійної роботи. Оцінити асимптотичну складність
алгоритму сортування злиттям, використовуючи головну теорему рекурсії.**

```python
def merge(left, right):
    # Зливає два відсортованих масиви left і right у один
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(list):
    # Рекурсивна функція сортування з використанням попередньої функції
    if len(list) < 2:
        return list
    middle = len(list) // 2
    left = mergesort(list[:middle])
    right = mergesort(list[middle:])
    return merge(left, right)

# Приклад сортування
a = [6, 5, 3, 1, 8, 7, 12, 9, 13]
print(f"Відсортований масив: {mergesort(a)}")
```

### 1. Надати визначення складності задачі із символом \( \Omega \).

**Розв'язок:**  
Символ \( \Omega \) використовується для позначення **найгіршої** (нижньої) межі складності алгоритму. Якщо для алгоритму має місце рівність:

$ f(n) = \Omega(g(n)$

це означає, що для достатньо великих значень \( n \), час виконання алгоритму не буде меншою за деяку константу, помножену на функцію g(n). 
Тобто $\Omega$ -нотація описує найгірший можливий час виконання.

**2. Функція часової складності має вигляд:**
$ F(N) = N^3 + 7N^2 - 14N $ Як записати асимптотичну складність у \( O \)-нотації?

**Розв'язок:**  
У функції $F(N) = N^3 + 7N^2 - 14N$
Асимптотична складність визначається терміном, який зростає найшвидше при збільшенні  N .  
Звісно, $N^3$ зростає найшвидше.  

**Отже, асимптотична складність:**

$ F(N) = O(N^3) $



### 3. Функція часової складності має вигляд:  
$ F(N) = 1.01N + N^{10}$. Як записати асимптотичну складність у \( O \)-нотації?

**Розв'язок:**  
В даному випадку найважливіший термін - це $N^{10} $, оскільки він має найбільшу ступінь. Тому асимптотична складність буде:

$O(N^{10})$

### 4. Функція часової складності має вигляд:  
$F(N) = N^{1.3} + 10 \log_2 N$
Як записати асимптотичну складність у $ O $-нотації?

**Розв'язок:**  
У цій функції основний термін - це $N^{1.3}$, оскільки він зростає швидше, ніж логарифмічний термін $10 \log_2 N$. Тому асимптотична складність буде:

$O(N^{1.3})$

### 6. Які існують способи підвищення обчислювальної швидкості алгоритмів? Який з них є найефективнішим?
 
Існує кілька способів підвищення обчислювальної швидкості алгоритмів:
1. *Оптимізація алгоритмів*: зменшення кількості операцій у кожному кроці або використання кращих структур даних.

3. *Кешування*: зберігання результатів обчислень для повторного використання, щоб уникнути зайвих операцій.
4. *Використання апаратного прискорення*: застосування спеціалізованих процесорів або графічних процесорів (GPU).

Найефективнішим методом є **паралелізація**, оскільки вона дозволяє значно прискорити обчислення при обробці великих обсягів даних.
