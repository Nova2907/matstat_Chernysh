− написати функцію pop_n(), що видаляє елементи стека з його початку до
номера n включно;
− оцінити асимптотичну складність (у середньому і в найгіршому разі)
процедур search, insert і delete роботи зі стеком.

```python
class Stack:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def size(self):
        return len(self.items)

    def pop_n(self, n):
        if n > len(self.items):
            raise IndexError("n перевищує розмір стека")
        removed_items = []
        for _ in range(n):
            removed_items.append(self.pop())
        return removed_items


# Тестування
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.push(6)
s.push(7)
print("Вміст стека до виконання pop_n:", s.items)
print("Видалені елементи:", s.pop_n(3))
print("Вміст стека після виконання pop_n:", s.items)
```

**Оцінка асимптотичної складності для стеку**
1. push (вставка): O(1), бо додається елемент у кінець списку.
2. pop (видалення): O(1), бо видаляється елемент із кінця списку.
3. search (пошук): O(n), бо в найгіршому разі доводиться переглянути весь стек.

**Написати функцію print_n(), що друкує елементи черги з його початку до номера n включно;**

```python
class Queue:
    def __init__(self):
        self.items = []  # Список для збереження елементів черги

    def isEmpty(self):
        return self.items == []  # Повертає True, якщо черга порожня

    def enqueue(self, item):
        self.items.insert(0, item)  # Додає елемент у кінець черги

    def dequeue(self):
        if self.isEmpty():
            raise IndexError("Спроба видалити елемент із порожньої черги")
        return self.items.pop()  # Видаляє елемент із початку черги

    def size(self):
        return len(self.items)  # Повертає кількість елементів у черзі

    def print_n(self, n):
        """Друкує перші n елементів черги."""
        if n > self.size():
            raise IndexError("n перевищує розмір черги")
        print("Перші", n, "елементів черги:", self.items[::-1][:n])


# Тестування
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
q.enqueue(6)
q.enqueue(7)

print("Вміст черги:", q.items[::-1])  # Показуємо чергу у звичному порядку
q.print_n(3)  # Друкуємо перші 3 елементи
```

**Оцінка асимптотичної складності**
Оцінка основних операцій черги:

1. **search (пошук):**

У середньому: O(n), бо потрібно переглянути всі елементи.

У найгіршому випадку: O(n), коли шуканий елемент знаходиться в кінці або відсутній.

2. **insert (enqueue):**

У середньому: O(n), оскільки новий елемент додається на початок списку, що викликає зміщення всіх інших елементів.

У найгіршому випадку: O(n), аналогічно.

3. **delete (dequeue):**

У середньому: O(1), оскільки видалення останнього елемента списку виконується за постійну кількість часу.

У найгіршому випадку: O(1), аналогічно.

#### **Контрольні питання**

**1. Що таке стек і які операції можна виконувати зі стеком?
Стек — структура даних за принципом LIFO (останній увійшов, перший вийшов). Операції:**

- push(item) — додати елемент.
- pop() — видалити верхній елемент.
- peek() — отримати верхній елемент без видалення.
- isEmpty() — перевірити, чи стек порожній.
- size() — отримати розмір стека.

**2. Яка основна відмінність між стеком та чергою?**

Стек: працює за принципом LIFO(перший увійшов, останій вийшов).

Черга: працює за принципом FIFO (перший увійшов, перший вийшов).

**3. Як ви можете реалізувати стек за допомогою масиву і за допомогою
зв’язаного списку?**

**За допомогою масиву:** Стек можна реалізувати за допомогою списку (масиву) в Python. Операція push (додавання елемента) здійснюється через метод append(), а pop (видалення елемента) — через метод pop(), який працює з кінця списку. Цей метод є швидким і має часову складність O(1). Однак, при фіксованому розмірі масиву може виникнути потреба в переналаштуванні (ресайзинг), що знижує ефективність при великих обсягах даних.

**за допомогою зв’язаного списку:** Зв'язаний список дозволяє створювати стек, де кожен елемент є окремим об'єктом (вузлом), що містить значення і посилання на наступний елемент. Операція push додає новий вузол на початок списку, а pop видаляє перший вузол. Це дає перевагу у вигляді динамічного розміру стека, оскільки кількість елементів не обмежена розміром масиву.



**Які переваги та недоліки кожного підходу?**

| Критерій               | Масив                                                                                       | Зв'язаний список                                                                            |
|------------------------|---------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------|
| **Розмір**             | Обмежений, потрібно перевизначати при переповненні.                                          | Динамічний, розмір змінюється автоматично.                                                 |
| **Часова складність**  | O(1) для операцій `push` та `pop`, але може бути O(n) для збільшення розміру масиву.        | O(1) для операцій `push` та `pop`, але доступ до елементів - O(n).                         |
| **Пам'ять**            | Вимагає пам'яті для самого масиву.                                                          | Вимагає додаткової пам'яті для кожного елемента (посилання на наступний).                 |
| **Складність реалізації** | Простий та зручний в реалізації.                                                              | Складніший, вимагає додаткової структури (вузлів) для кожного елемента.                    |
| **Гнучкість**          | Менш гнучкий (фіксований розмір).                                                           | Більш гнучкий (динамічно змінюється розмір).                                                |
