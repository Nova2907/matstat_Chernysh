3. Виконати теоретичні розрахунки асимптотичної складності алгоритмів
сортування методами вставляння та бульбашки.

Алгоритм вставляння (Insertion Sort):
Алгоритм сортування методом вставляння працює шляхом поетапного перебору елементів і вставлення їх у правильне місце серед відсортованої частини масиву.

 1)Найгірший випадок: O(n²) (масив у зворотному порядку).
 
 2)Найкращий випадок: O(n) (масив вже відсортований).

Пояснення: В найгіршому випадку для кожного елемента доведеться порівнювати з усіма попередніми елементами.

Алгоритм бульбашки (Bubble Sort):
Алгоритм сортування методом бульбашки працює шляхом порівняння сусідніх елементів масиву і їх обміну, якщо вони не відсортовані. Процес повторюється кілька разів до тих пір, поки весь масив не буде відсортовано.

 1)Найгірший випадок: O(n²) (масив у зворотному порядку).
 
 2)Найкращий випадок: O(n) (масив вже відсортований).

Пояснення: У найгіршому випадку потрібно пройти по масиву n разів, порівнюючи і обмінюючи елементи.

4. Реалізувати на Python алгоритм сортування методом бульбашки.

```python
# Алгоритм сортування методом бульбашки
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):  
            if arr[j] > arr[j+1]:  
                arr[j], arr[j+1] = arr[j+1], arr[j]  
    return arr

# Приклад масиву для сортування
array = [69, 52, 7, 13, 12, 1, 99, 9]

print("Несортований масив:", array)

sorted_array = bubble_sort(array)

print("Відсортований масив:", sorted_array)
```

5. Дослідити реальний час виконання обох алгоритмів на Python і
побудувати графіки залежностей Tinsert(n) та Tbuble(n) для n =
[5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000],
згенерувавши відповідні цілочисельні масиви.

```python
import time
import random
import matplotlib.pyplot as plt

# сортування методом вставляння
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# сортування методом бульбашки
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Генерація масиву та вимірювання часу
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000]
insertion_times = []
bubble_times = []

for size in sizes:
    test_array = [random.randint(0, 10000) for _ in range(size)]

    # Час для insertion sort
    array_copy = test_array[:]
    start_time = time.time()
    insertion_sort(array_copy)
    insertion_times.append(time.time() - start_time)

    # Час для bubble sort
    array_copy = test_array[:]
    start_time = time.time()
    bubble_sort(array_copy)
    bubble_times.append(time.time() - start_time)

plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_times, label="метод вставлення", marker='o')
plt.plot(sizes, bubble_times, label="метод бульбашки", marker='o')
plt.title("Час виконання алгоритмів сортування")
plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час виконання (секунди)")
plt.legend()
plt.grid()
plt.show()
```

6. Надати відповіді на контрольні питання.

1) O-нотація та Ω-нотація:

O-нотація: Описує верхню межу складності алгоритму, тобто, максимальний час або кількість операцій, яку алгоритм може виконати для вхідних даних розміру N.
Ω-нотація: Описує нижню межу складності алгоритму, тобто, мінімальний час або кількість операцій, яку алгоритм буде виконувати для найкращих умов.
Відмінність: O-нотація визначає найгірший випадок, тоді як Ω-нотація визначає найкращий випадок.
Часова складність алгоритму сортування методом вставляння за найсприятливіших умов:

2) Найсприятливіші умови: масив уже відсортований.
Часова складність: O(n). У найкращому випадку кожен елемент буде вставлятися лише один раз без необхідності пересування інших елементів.
Ефективний алгоритм:

3) Ефективним називається алгоритм, який для певного розміру вхідних даних виконується за мінімальний час, а також використовує оптимальну кількість пам'яті. Оцінка ефективності часто включає як часову складність, так і використання пам'яті.
Головний параметр (розмір) завдання:

4) Головний параметр завдання — це змінна, яка визначає розмір вхідних даних для алгоритму, наприклад, кількість елементів у масиві для алгоритмів сортування. Розмір завдання впливає на часову складність алгоритму.

5) Асимптотична складність функції F(N) = N³ + 7N² − 14N:
Основний член, який визначає асимптотику, — це найвищий степінь N.
Оскільки найвищий степінь у функції — N³, асимптотична складність у O-нотації: O(N³).
